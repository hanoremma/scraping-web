```python
df = df.dropna(subset=["Harga"])

df = df[df["Harga"] != "Mencari"]

# Hapus harga sewa
df = df[
    ~df["Harga"].str.contains(
        "bulan|tahun|hari",
        case=False,
        na=False
    )
]
```

```python
def convert_price(x):

    if pd.isna(x):
        return np.nan

    x = str(x).strip()

    # Harus diawali Rp
    if not x.startswith("Rp"):
        return np.nan

    # Hilangkan Rp
    x = x.replace("Rp", "").strip()

    # Ambil angka + satuan
    match = re.match(r"([\d.,]+)\s*(Rb|Jt|M)?", x)

    if not match:
        return np.nan

    angka = match.group(1).replace(",", ".")
    satuan = match.group(2)

    try:
        angka = float(angka)
    except ValueError:
        return np.nan

    if satuan == "Rb":
        return angka * 1_000

    elif satuan == "Jt":
        return angka * 1_000_000

    elif satuan == "M":
        return angka * 1_000_000_000

    else:
        return angka
```

```python
df["Harga"] = df["Harga"].apply(convert_price)
```

```python
def convert_area(x):

    if pd.isna(x):
        return np.nan

    angka = re.findall(r"\d+", x)

    if angka:
        return int(angka[0])

    return np.nan
```

```python
df["Luas Total"] = df["Luas Total"].apply(convert_area)
```

```python
def convert_number(x):

    if pd.isna(x):
        return np.nan

    # Jika sudah angka, langsung kembalikan
    if isinstance(x, (int, float)):
        return x

    x = str(x)

    angka = re.findall(r"\d+", x)

    if angka:
        return int(angka[0])

    return np.nan
```

```python
df["Kamar Tidur"] = df["Kamar Tidur"].apply(convert_number)
df["Kamar Mandi"] = df["Kamar Mandi"].apply(convert_number)
```
