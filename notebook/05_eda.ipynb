```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("whitegrid")

df = pd.read_csv("data/lamudi_apartemen_clean_final.csv")
```

```python
numerik = [
    "Harga",
    "Luas Total",
    "Kamar Tidur",
    "Kamar Mandi"
]

for kolom in numerik:

    plt.figure(figsize=(8,4))

    sns.histplot(df[kolom], bins=30, kde=True)

    plt.title(f"Distribusi {kolom}")

    plt.show()
```

```python
for kolom in numerik:

    plt.figure(figsize=(8,2))

    sns.boxplot(x=df[kolom])

    plt.title(f"Boxplot {kolom}")

    plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x="Luas Total",
    y="Harga"
)

plt.title("Hubungan Luas Total terhadap Harga")

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x="Kamar Tidur",
    y="Harga"
)

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x="Kamar Mandi",
    y="Harga"
)

plt.show()
```

```python
corr = df[[
    "Harga",
    "Luas Total",
    "Kamar Tidur",
    "Kamar Mandi"
]].corr()

plt.figure(figsize=(6,5))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Heatmap Korelasi")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(
    data=df,
    y="Kota",
    order=df["Kota"].value_counts().head(10).index
)

plt.title("Top 10 Kota")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    y="Provinsi",
    order=df["Provinsi"].value_counts().index
)

plt.title("Distribusi Provinsi")

plt.show()
```

```python
rata_harga = (
    df.groupby("Kota")["Harga"]
      .mean()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(10,5))

sns.barplot(
    x=rata_harga.values,
    y=rata_harga.index
)

plt.title("Rata-rata Harga per Kota")

plt.show()
```
